## Load up HouseX ground-truth embeddings

```python
import pickle

with open("/kaggle/input/housex-audio-embeds-experiments/audio-embeds/audio-embeds/ground-truth/clap_fused_housex_saved_embeds_dict.pkl", "rb") as f:
    gt_embeds = pickle.load(f)
```

```python
gt_embeds.keys()
```

## Method to calculate dot product between two embeddings

```python
def dot_product(u, v):
    x = u@v.T
    return float(x.detach().cpu())
```

```python
dot_product(gt_embeds['bass_house'][0], gt_embeds['bass_house'][0])
```

## Experiment Details
- Variant: raw
- Steps: 200

```python
from tqdm import tqdm

file_path = "/kaggle/input/housex-audio-embeds-experiments/audio-embeds/audio-embeds/custom-songs-embeds/embeddings_audioldm_raw_200_10_examples.pkl"
with open(file_path, "rb") as f:
    target_embeds = pickle.load(f)
    
all_keys_gt = list(gt_embeds.keys())
all_keys_target = list(target_embeds.keys())

assert sorted(all_keys_gt) == sorted(all_keys_target)

scoring_dict = {}
for k in all_keys_gt:
    if k not in scoring_dict:
        scoring_dict[k] = 0.0
        num_entries_computed = 0
        running_score = 0.0
        
        for u in tqdm(gt_embeds[k]):
            for v in target_embeds[k]:
                score = dot_product(u, v)
                running_score += score
                num_entries_computed += 1
        
        average_score = running_score/num_entries_computed
        print(f"Genre: {k}, Total score {running_score} over {num_entries_computed} computations")
        scoring_dict[k] = average_score
        
print("Variant: raw, 200")
print(scoring_dict)
```
